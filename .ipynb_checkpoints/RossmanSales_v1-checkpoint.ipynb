### Understanding Data - Rossmann Sales Forecasting

Rossmann operates over 3,000 drug stores in 7 European countries. Currently, Rossmann store managers are tasked with predicting their daily sales for up to six weeks in advance. Store sales are influenced by many factors, including promotions, competition, school and state holidays, seasonality, and locality. With thousands of individual managers predicting sales based on their unique circumstances, the accuracy of results can be quite varied

Submissions are evaluated on the Root Mean Square Percentage Error (RMSPE). The RMSPE is calculated as

<img src="images/rmspe.png" />
where y_i denotes the sales of a single store on a single day and yhat_i denotes the corresponding prediction. Any day and store with 0 sales is ignored in scoring.<br>

https://www.kaggle.com/c/rossmann-store-sales/overview

### Feature Engineering

```python
# feature creation (code based on  https://www.kaggle.com/cast42/xgboost-in-python-with-rmspe-v2/code)
def create_features(data):
    
    mappings = {'0':0, 'a':1, 'b':2, 'c':3, 'd':4}
    data.StoreType.replace(mappings, inplace=True)
    data.Assortment.replace(mappings, inplace=True)
    data.StateHoliday.replace(mappings, inplace=True)
    
    
    data['Year'] = data.Date.dt.year
    data['Month'] = data.Date.dt.month
    data['Day'] = data.Date.dt.day
    data['DayOfWeek'] = data.Date.dt.dayofweek
    data['WeekOfYear'] = data.Date.dt.weekofyear
    
    
    ##Converting respective columns to noof months from given record date
    data['CompetitionOpen'] = 12 * (data.Year - data.CompetitionOpenSinceYear) + (data.Month - data.CompetitionOpenSinceMonth)
    data['PromoOpen'] = 12 * (data.Year - data.Promo2SinceYear) + (data.WeekOfYear - data.Promo2SinceWeek) / 4.0
    
    data['CompetitionOpen'] = data.CompetitionOpen.apply(lambda x: x if x > 0 else 0)        
    data['PromoOpen'] = data.PromoOpen.apply(lambda x: x if x > 0 else 0)
    
  
    month2str = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sept', 10:'Oct', 11:'Nov', 12:'Dec'}
    data['monthStr'] = data.Month.map(month2str)
    data.loc[data.PromoInterval == 0, 'PromoInterval'] = ''
    data['IsPromoMonth'] = 0
    for interval in data.PromoInterval.unique():
        if interval != '':
            for month in interval.split(','):
                data.loc[(data.monthStr == month) & (data.PromoInterval == interval), 'IsPromoMonth'] = 1

    return data
```

```python
create_features(train_data)
create_features(val_data)
create_features(test)
print('Features creation finished')
```

#### Defining Loss function -- RMSPE & Early Stopping Criteria

```python
def rmspe(y, yhat):
    return np.sqrt(np.mean(np.square(1-(yhat/y))))

def rmspe_NN(y, yhat):
    y = np.expm1(y)
    yhat = np.expm1(yhat)
    return rmspe(y,yhat)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, CSVLogger, Callback

class RMSPE_callback(Callback):
    def __init__(self,training_data,validation_data):
        self.x = training_data[0]
        self.y = training_data[1]
        self.x_val = validation_data[0]
        self.y_val = validation_data[1]

    def on_train_begin(self, logs={}):
        return

    def on_train_end(self, logs={}):
        return

    def on_epoch_begin(self, epoch, logs={}):
        return

    def on_epoch_end(self, epoch, logs={}):
        y_pred = self.model.predict(self.x, verbose=0)
        logs['rmspe'] = rmspe_NN(self.y, y_pred)

        y_pred_val = self.model.predict(self.x_val, verbose=0)
        logs['val_rmspe'] = rmspe_NN(self.y_val, y_pred_val)

        print('\rTrain_RMSPE: %s - Val_RMSPE: %s' % (str(round(logs['rmspe'],5)),str(round(logs['val_rmspe'],5)))+'\n')
        return

    def on_batch_begin(self, batch, logs={}):
        return

    def on_batch_end(self, batch, logs={}):
        return
```

```python
early_stopping = EarlyStopping(monitor='val_root_mean_squared_error', patience=5, mode='min', verbose=1)
    
model_checkpoint = ModelCheckpoint(
        os.path.join(save_dir, 'custom_keras_model_v1'),
        monitor='val_root_mean_squared_error', mode='min', # mode must be set to max or Keras will be confused
        save_best_only=True,
        verbose=0)
```

```python
callbacks = [early_stopping, model_checkpoint] #RMSPE_callback(training_data=(X_train_arr, y_train),validation_data=(X_val_arr, y_val)), 
```

### Building the NN Model

```python
def buildSimpleNN(X_train, activation):
    
    model = Sequential()
    model.add(Dense(32, input_dim= X_train.shape[1], kernel_initializer='glorot_normal',))
    model.add(Activation(activation))
    
    model.add(Dropout(0.1))
    
    model.add(Dense(64,  kernel_regularizer=l1(0.001), bias_regularizer=l1(0.001)))
    model.add(Activation(activation)) 
    model.add(Dropout(0.2))
    
    model.add(Dense(128, kernel_regularizer=l1(0.01), bias_regularizer=l1(0.01))) #
    model.add(Activation(activation))
   
    
    model.add(Dense(32,)) #
    model.add(Activation(activation))

    model.add(Dense(1, activation='linear'))
    return model
```

```python
model_struc_data = buildSimpleNN(X_train_arr, 'relu')
model_struc_data.summary()
```

```python
# initiate Adam optimizer
opt = Adam(learning_rate=0.001, beta_1=0.9,beta_2=0.999) #Using defaults

# Let's train the model using Adam
model_struc_data.compile(loss='mse', optimizer=opt, metrics=[tf.keras.metrics.RootMeanSquaredError()])
```

```python
train_output = model_struc_data.fit(X_train_arr, y_train, epochs=100, 
                                    batch_size=batch_size, validation_data=(X_val_arr,y_val),callbacks=callbacks)
```

```python
yhat_train = model_struc_data.predict(X_train_arr)
yhat_val = model_struc_data.predict(X_val_arr)
```

```python
res = pd.DataFrame(data = y_val)
res['Prediction']=yhat_val
res = pd.merge(X_val,res, left_index= True, right_index=True)
res['Ratio'] = res.Prediction/res.Sales
res['Error'] =abs(res.Ratio-1)
res['Weight'] = res.Sales/res.Prediction
res.head()
```

```python
col_1 = ['Sales','Prediction']
col_2 = ['Ratio']
L=np.random.randint( low=1,high = 1115, size = 3 ) 
print('Mean Ratio of prediction and real sales data is {}: store all'.format(res.Ratio.mean()))
for i in L:
    
    s1 = pd.DataFrame(res[res['Store']==i],columns = col_1)
    s2 = pd.DataFrame(res[res['Store']==i],columns = col_2)
    s1.plot(title = 'Comparation of predition and real sales data: store {}'.format(i),figsize=(12,4))
    s2.plot(title = 'Ratio of prediction and real sales data: store {}'.format(i),figsize=(12,4))
    print('Mean Ratio of predition and real sales data is {}: store {}'.format(s2.Ratio.mean(),i))
```

```python
# whole correction
print("weight correction")
W=[(0.95+(i/1000)) for i in range(40)]
S =[]
for w in W:
    error = rmspe(np.expm1(y_val), np.expm1(yhat_val*w).reshape(-1,))
    print('RMSPE for {:.3f}:{:.6f}'.format(w,error))
    S.append(error)
Score = pd.Series(S,index=W)
Score.plot()
BS = Score[Score.values == Score.values.min()]
print ('Best weight for Score:{}'.format(BS))
```

```python
# correction by store
L=range(1115)
W_val=[]
W_test=[]
for i in L:
    s1 = pd.DataFrame(res[res['Store']==i+1],columns = col_1)
    s2 = pd.DataFrame(X_test[X_test['Store']==i+1])
    W1=[(0.95+(i/1000)) for i in range(40)]
    S =[]
    for w in W1:
        error = rmspe(np.expm1(s1.Sales), np.expm1(s1.Prediction*w))
        S.append(error)
    Score = pd.Series(S,index=W1)
    BS = Score[Score.values == Score.values.min()]
    a=np.array(BS.index.values)
#     print(f"Best Score is {BS}")
#     print(f"a = {a}")
    b_val=a.repeat(len(s1))
    b_test=a.repeat(len(s2))
    W_val.extend(b_val.tolist())
    W_test.extend(b_test.tolist())
```

```python
print(len(W_val))
print(yhat_val.shape)
print(y_val.shape)

yhat_new_val = yhat_val.reshape(-1,)*W_val
error = rmspe(np.expm1(y_val), np.expm1(yhat_new_val))
print ('RMSPE for weight corretion {:6f}'.format(error))
```

```python
yhat_new_val = yhat_val.reshape(-1,)*0.976
error = rmspe(np.expm1(y_val), np.expm1(yhat_new_val))
print ('RMSPE for weight corretion {:6f}'.format(error))
```
